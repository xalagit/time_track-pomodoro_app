```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear grafo dirigido
G = nx.DiGraph()

# Nombres de los mÃ³dulos
modules = ["main.py", "gui.py", "timer.py", "database.py", "stats.py"]
G.add_nodes_from(modules)

# Conexiones entre archivos
edges = [
    ("main.py", "database.py"),
    ("main.py", "gui.py"),
    ("gui.py", "timer.py"),
    ("gui.py", "stats.py"),
    ("timer.py", "database.py")
]

G.add_edges_from(edges)

# Layout visual
pos = nx.spring_layout(G, seed=42)

# Dibujar grafo
plt.figure(figsize=(10, 6))
nx.draw_networkx_nodes(G, pos, node_color="#66b3ff", node_size=2500)
nx.draw_networkx_edges(G, pos, edge_color="gray", arrows=True, arrowsize=25)
nx.draw_networkx_labels(G, pos, font_size=12, font_family="monospace")

plt.title("ðŸ“Š Diagrama de interacciÃ³n entre mÃ³dulos de la Pomodoro App", fontsize=14)
plt.axis("off")
plt.tight_layout()
plt.show()
```
